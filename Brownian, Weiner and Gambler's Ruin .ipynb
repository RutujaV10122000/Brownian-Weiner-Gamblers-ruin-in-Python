```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
T = 1.
times = np.linspace(0.,T,n)
print(times)
```

```python
dt = times[1]-times[0]
# W(t) ~ Normal with mean 0 and variance 1.
dW =np.sqrt(dt) * np.random.normal(size=(n-1,)) 
W0 = np.zeros(shape=(1,))   
W = np.concatenate((W0, np.cumsum(dW,)))
plt.plot(times, W)
plt.show()
```

```python
n = 10000
d = 5
T = 1.
times = np.linspace(0.,T,n)
print(times)
```

```python
dt = times[1]-times[0]
# W(t2) - W(t1) ~ Normal with mean 0 and variance t2-t1.
dW =np.sqrt(dt) * np.random.normal(size=(n-1, d)) 
W0 = np.zeros(shape=(1, d))   
W = np.concatenate((W0, np.cumsum(dW, axis = 0)), axis=0)
plt.plot(times, W)
plt.show()
```

```python
# how many tosses until we go broke ?
import random
random.random()
```

```python
def integer_walk():
    money, count = 15,0
    while money:
        toss = random.random()
        if toss < 0.50:
            money -= 5
        if toss > 0.50:
            money += 5
        count += 1
    return count
```

```python
# Simulation 1000
sims = [integer_walk() for x in range(1000)]
print(len(sims))
```

```python
# Number of times coin rolled out before you ran out of money
sims
```
